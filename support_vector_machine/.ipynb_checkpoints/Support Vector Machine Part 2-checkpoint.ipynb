```python
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np

from sklearn.svm import SVC
from EML_plot import plot_svc, ilustrasi_RBF
from ipywidgets import interact, fixed

df = read_csv("./data2.csv")
df.tail(15).T
```

```python
X = df.drop("Label",axis=1)
y = df.Label
```

```python
plt.figure(figsize=(4,4))
plt.scatter(X.x_1, X.x_2, c=y, cmap="bwr", s=5)
```

C in SVM is for regularization

```python
# model = SVC(kernel="linear",  C=1e10)
# model.fit(X,y)
# plot_svc(X,y,model)
```

```python
model = SVC(kernel="rbf", C=1e10)
model.fit(X,y)
plot_svc(X,y,model)
```

# Kernelized SVM

```python
ilustrasi_RBF(X,y)
interact(ilustrasi_RBF, X=fixed(X), y=fixed(y), sigma=2)
```
