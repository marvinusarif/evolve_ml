```python
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np

from sklearn.svm import SVC
from EML_plot import plot_svc, ilustrasi_RBF
from ipywidgets import interact, fixed

df = read_csv("./data3.csv")
df.tail(15).T
```

```python
X = df.drop("Label",axis=1)
y = df.Label
```

```python
plt.figure(figsize=(4,4))
plt.scatter(X.x_1, X.x_2, c=y, cmap="bwr", s=5)
```

C in SVM is for regularization

When C is bigger then there will be smaller tolerance
When C is smaller then there will be larger tolerance

When Gamma is larger then the data will be sharper
When Gamma is smaller then the data distribution will be more distributed

```python
model = SVC(kernel="rbf", gamma=20, C=1e3)
model.fit(X,y)
plot_svc(X,y,model, support=False)
```

# Kernelized SVM

```python
ilustrasi_RBF(X,y)
interact(ilustrasi_RBF, X=fixed(X), y=fixed(y), sigma=2)
```
