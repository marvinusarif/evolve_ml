```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sknn.mlp import Layer, Classifier
from random import seed
import datetime

def create_network(niter, lr, verboseflag) :
    nn = Classifier(
    layers=[
        Layer("Rectifier", units=5), #this is 1st hidden layer,
        Layer("Rectifier", units=3), #this is 2nd hidden layer,
        Layer("Sigmoid")], #last layer will be number of output - we expect 1 number of output
        learning_rate=lr,
        learning_rule="rmsprop",
        n_iter=niter, verbose=verboseflag, loss_type='mcc')
    return nn

def accuracy_metric(actual, predicted) :
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            correct += 1
    return correct / float(len(actual)) * 100.0


# prepare data
if __name__ == "__main__":
    seed(datetime.datetime.utcnow())
    x = np.genfromtxt('../knn/iris.data', delimiter=",", usecols=(0,1,2,3))
    y = np.genfromtxt('../knn/iris.data', delimiter=",", usecols=(4), dtype=str)
    
    # integer encode
    label_encoder = LabelEncoder()
    y_integer_encoded = label_encoder.fit_transform(y)
#     print(y_integer_encoded)
    
    # binary encode
    onehot_encoder = OneHotEncoder(sparse=False)
    y_integer_encoded = y_integer_encoded.reshape(len(y_integer_encoded), 1)
    y_onehot_encoded = onehot_encoder.fit_transform(y_integer_encoded)
#     print(y_onehot_encoded)
    
    niter = 100
    lr = 0.01
    verboseflag = True
    model = create_network(niter, lr, verboseflag)
    model.fit(x, y_onehot_encoded)
    
    predicted_train_y = model.predict(x, y_onehot_encoded)
    print(predicted_train_y)
# #     acc_train = accuracy_metric(y_onehot_encoded, predicted_train_y)
#     print('accuracy on train set', acc_train)
    
```
